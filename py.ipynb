```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

conv_layer = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)

weights = conv_layer.weight.detach()

fig, ax = plt.subplots(1, 3, figsize=(8, 3))
for i in range(3):
    ax[i].imshow(weights[0, i, :, :], cmap='gray')
    ax[i].set_title(f'Channel {i+1}')
    ax[i].axis('off')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

conv_layer = tf.keras.layers.Conv2D(filters=16, kernel_size=3, activation='relu', input_shape=(32, 32, 3), padding='same')

model = tf.keras.Sequential([conv_layer])

weights = conv_layer.get_weights()[0]

num_filters = weights.shape[3]
input_channels = weights.shape[2]

fig, axes = plt.subplots(num_filters, input_channels, figsize=(2 * input_channels, 2 * num_filters))

for i in range(num_filters):
    for j in range(input_channels):
        ax = axes[i, j]
        ax.imshow(weights[:, :, j, i], cmap='gray')
        ax.set_title(f'F{i+1}, C{j+1}', fontsize=8)
        ax.axis('off')

plt.tight_layout()
plt.show()
```
